## Total Sales by Region

```python
sns.set_style('whitegrid')
sns.set_context('poster',font_scale=0.5)
sns.barplot(x='Region',y='Sales_Amount',data=df,palette='rainbow')
```

## Sales Trend Over Time

```python
df['Sale_Date']=pd.to_datetime(df['Sale_Date'])

Monthly_sale=df.groupby(df['Sale_Date'])['Sales_Amount'].sum()
Monthly_sale
```

```python
plt.figure(figsize=(10,6))
plt.style.use('ggplot')
Monthly_sale.plot.line()
plt.xlabel('Sale Date')
plt.ylabel('Sale Amount')
plt.title('Monthly Wise sales')
```

## Sales Amount by Customer Type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Customer_Type',y='Sales_Amount',data=df,hue='Region')
plt.legend(loc=1)
```

## Top Sales Reps

```python
S=sns.barplot(x='Sales_Rep',y='Sales_Amount',data=df,palette='rainbow')
S.set_title('Top Sale Reps')
```

## Sales_Channel Wise Sales

```python
S=sns.barplot(x='Sales_Channel',y='Sales_Amount',data=df,palette='rainbow')
S.set_title('Sales Channel Wise Sales')
```

## Product_Category wise Sales

```python
plt.figure(figsize=(10,6))
P=sns.barplot(x='Product_Category',y='Sales_Amount',data=df,hue='Region',palette='rainbow')
P.set_title('Product Category Wise Sales')
plt.legend(loc=1)
```

## Product Wise Discount

```python
sns.barplot(x='Product_Category',y='Discount',data=df,palette='rainbow')
```

## Headmap of Total Sales Amount by Region and Sales Representative

```python
pivot=df.pivot_table(index='Region', columns='Sales_Rep',values='Sales_Amount')
pivot
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pivot,annot=True,linewidths=1)
```
