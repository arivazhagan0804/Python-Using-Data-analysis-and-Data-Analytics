## Year-Wise Rent Amount in India

```python
India=df[df['Country']=='India']
India
```

```python
plt.figure(figsize=(10,6))
plt.style.use('ggplot')
India.plot.bar(x='Year',y='Rent Index')
plt.xlabel('Year')
plt.ylabel('Rent')
plt.title('Year-Wise Rent Amount(India)')
plt.show()
```

## Rent Amount in All Country

```python
rent=df.groupby('Country')['Rent Index'].mean().sort_values()
rent
```

```python
plt.figure(figsize=(15,6))
sns.barplot(rent)
plt.show()
```

## Year-wise House Price Index

```python
price=df.groupby('Year')['House Price Index'].mean()
price
```

```python
plt.figure(figsize=(10,6))
plt.style.use('ggplot')
price.plot.line(marker='o',markerfacecolor='yellow')
plt.xlabel('price')
plt.ylabel('House Price Index')
plt.title('Year-wise House Price Index')
plt.legend()
plt.show()
```

## Year- wise GDP Growth

```python
year=df.groupby('Year')['GDP Growth (%)'].mean()
year
```

```python
plt.figure(figsize=(10,6))
plt.style.use('ggplot')
year.plot.line(marker='o',markerfacecolor='yellow')
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('Year-wise GDP Growth')
plt.legend()
plt.show()
```

```python
population=df.groupby('Country')['Population Growth (%)'].mean().sort_values()
population
```

```python
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
f=sns.barplot(population)
f.set(xlabel='Country',
      ylabel='Population Growth',
      title='Country-wise Population Growth')

```

```python
pivot=df.pivot_table(columns='Year',index='Country',values='Rent Index')
pivot
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(data=pivot,annot=True,linewidths=1)
plt.show()
```
