Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
df=online_rt[online_rt['Country']!='United Kingdom']
df
```

```python
Quantity_10=df.groupby('Quantity')['Country'].max()
most=Quantity_10.tail(30)
most
plt.figure(figsize=(15,6))
most.hist(bins=10)
plt.xlabel('country')
plt.ylabel('Quantity')
plt.title('Most Quantity order')
plt.show()
```

Step 5 : Exclude negative Quantity entries

```python
Negative=Quantity_10.head(20)
plt.figure(figsize=(10,6))
Negative.hist()
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.title('Negative Quantity')
plt.show()
```

 Step 12. Group by CustomerID and Country and find out the average price (AvgPrice) each customer spends per unit.

```python
AvgPrice_CustomerID=online_rt.groupby('CustomerID')['UnitPrice'].mean()
AvgPrice_CustomerID
```
