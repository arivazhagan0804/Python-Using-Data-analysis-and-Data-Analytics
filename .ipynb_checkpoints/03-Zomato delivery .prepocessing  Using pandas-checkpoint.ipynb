Grouping rows by time

```python
df['Order Placed At']=pd.to_datetime(df['Order Placed At'])
df['Order Placed At']
```

**Applying function to create 2 columns called year,month**

```python
df['year']=df['Order Placed At'].apply(lambda x:x.year)
df['year']
```

```python
df.groupby(df['year'])['Rating'].mean()
```

```python
df['month']=df['Order Placed At'].apply(lambda x:x.month)
df['month'].sort_values()
```

**Create Month wise Total Sales**

```python
df['month'].unique()
```

```python
df.groupby(df['month'])['Total'].mean()
```

**Create City wise Total sales**

```python
df['City'].unique()
```

**Create subzone wise Total sales**

```python
df.groupby('Subzone')['Total'].mean()
```

**Create Restaurant Wise Total Sales**

```python
df.groupby('Restaurant name')['Total'].sum()
```
