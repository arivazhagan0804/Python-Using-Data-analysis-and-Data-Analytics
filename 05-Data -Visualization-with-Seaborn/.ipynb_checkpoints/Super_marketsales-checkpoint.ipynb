## Branch-Wise total Bill

```python
sns.set_style('whitegrid')
sns.set_context('poster',font_scale=0.5)
sns.barplot(x='Branch',y='Total',data=df,palette='rainbow')
plt.show()
```

head()-It will give us to first five index in your dataset.

```python
df.head()
```

## Distribution of customers by City

```python
plt.figure(figsize=(20,6))
sns.displot(df['City'],palette='rainbow')
plt.show()
```

## Linear Relationship Between Total and Gross Income

```python
sns.lmplot(x='Total',y='gross income',hue='Gender',data=df)
plt.show()
```

## Total Bill amount Distribution by Genter

```python
sns.barplot(x='Gender',y='Total',data=df,hue='City',palette='rainbow')
plt.legend(loc=1)
plt.show()
```

```python
df.head()
```

## Total Bill by Product

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Product line',y='Total',data=df,palette='rainbow')
plt.show()
```

```python
df.head()
```

# Number of Payment Transactions per Category

```python
sns.countplot(x='Payment',data=df,palette='rainbow')
plt.show()
```

## Box plot of Grocery Purchase Totals

```python
sns.boxplot(x='Gender',y='Total',palette='rainbow',data=df)
plt.show()
```

```python
sns.violinplot(x='Branch',y='Total',hue='Gender',data=df)
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Product line',y='Rating',palette='rainbow',data=df)
plt.show()
```

```python
corr=df[['Total','gross income']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,linewidths=1,linecolor='yellow')
plt.show()
```
