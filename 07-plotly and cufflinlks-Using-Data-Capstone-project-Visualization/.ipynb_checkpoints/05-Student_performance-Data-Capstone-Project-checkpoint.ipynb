## Create New column Total Score

```python
total_score=df['math score']+df['reading score']+df['writing score']
total_score
```

```python
df['total score']=total_score
df
```

## Average Total Score by Parental Level of Education

```python
fig9=px.bar(df,x='parental level of education',y='total score')
fig9.show()
```

## 3d_line plot

```python
fig10=px.line_3d(df,x='race/ethnicity',y='parental level of education',z='gender',color='total score')
fig10.show()
```

## Create pivot  table

```python
pivot_table=df.pivot_table(index='parental level of education',columns='race/ethnicity',values='total score')
pivot_table
```

##  Headmap of total Score by Race and Parental Level of Education
