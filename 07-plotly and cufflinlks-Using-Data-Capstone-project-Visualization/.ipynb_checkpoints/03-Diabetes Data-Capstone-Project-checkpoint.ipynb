## Create Histogram age wise count

```python
df['Age'].iplot(kind='histogram',bins=10,xTitle='Age',yTitle='count',title='Distribution of Patients by Age')
```

## Create Scatter Age vs Pregnancies

```python
df.iplot(kind='scatter',x='Age',y='Pregnancies',mode='markers',title='Age wise pregnancies',xTitle='Age',yTitle='Pregnancies')
```

## create Scatter Age vs Glucose

```python
df.iplot(kind='scatter',x='Age',y='Glucose',title='Age vs Glucose Scatter',mode='markers',asFigur=True,xTitle='Age',yTitle='Glucose_range')
```

## Create Scatter Age vs BloodPressure 

```python
df.iplot(kind='scatter',x='Age',y='BloodPressure (mg/dL)',mode='markers',xTitle='Age',yTitle='BloodPressure',title='Age vs BloodPressure')
```
